```python
import pandas as pd
import re
```

```python
pd.set_option('display.max_colwidth', None)
column_names = ['target', 'ids', 'date', 'flag', 'user', 'text']

df = pd.read_csv('../Data/Twitter/data.csv', encoding = 'latin1', header=None, names=column_names)
df
```

```python
drop_columns = ['ids', 'date', 'flag', 'user']
df.drop(columns=drop_columns, axis=1, inplace=True)
df.rename(columns={'target':'label'}, inplace=True)
df['label'].replace({4:1},inplace=True)
df
```

```python
# Remove urls
url_pattern = re.compile(r'https?://\S+|www\.\S+|\b\w+\.com\b|\b\w+\.org\b')
df['text'] = df['text'].apply(lambda x: url_pattern.sub('', x))
df
```

```python
#removing usernames
username_pattern = re.compile(r'@\w+')
df['text'] = df['text'].apply(lambda x: username_pattern.sub('', x))
df
```

```python
df_small = df.groupby('label', group_keys=False).apply(lambda x: x.sample(10000))
```

```python
df_small.to_csv('../Data/Twitter/cleaned_data.csv', index=False)
df_small['label'].value_counts
```
